```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
```

```python
data = pd.read_csv('FitBit data.csv')
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data1 = data.copy()
```

```python
data1['ActivityDate'].unique()
```

```python
data1['ActivityDate'].head(10)
```

```python
# Adding Date, Month and year to the dataset

data1['year'] = pd.DatetimeIndex(data1['ActivityDate']).year
data1['month'] = pd.DatetimeIndex(data1['ActivityDate']).month
data1['date'] = pd.DatetimeIndex(data1['ActivityDate']).date
```

```python
data1.head(10)
```

```python
# Drop the columns

data1 = data1.drop(['TotalDistance'],axis = 1)
data1.head(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,8))

ax = sns.boxplot(x='date', y = 'Calories', data = data1)
ax = sns.swarmplot(x='date', y = 'Calories', data = data1, color="grey")

ax.set_title('Box plot of Calories')
```

```python
# converting the datatype to datetime

data1['Week'] = pd.to_datetime(data1.ActivityDate).dt.week
data1['Year'] = pd.to_datetime(data1.ActivityDate).dt.year
```

```python
data1.head()
```

```python
data1.ActivityDate.dtype
```

```python
data1['ActivityDate'] = pd.to_datetime(data1['ActivityDate'])
data1['day'] = data1['ActivityDate'].dt.weekday_name
```

```python
data1.head(10)
```

```python
plt.figure(figsize = (15,8))

ax = sns.barplot(x='date', y='Calories', data = data1)
ax.set_title('Barplot of calories by the day of the week')
```
